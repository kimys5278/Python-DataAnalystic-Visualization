# 데이터 수정 

```python
import pandas as pd 
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고'}) # 딕셔너리로 이름 바꿈
```

```python
df['학교'].replace({'북산고':'상북고','능남고':'화순고'})
```

```python
df
```

```python
df['학교'].replace({'북산고':'상북고','능남고':'화순고'},inplace=True)
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] +'등학교' #학교데이터에 '등학교'를 넣어서 수정
df
```

## Column추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학']+ df['과학']+ df['사회']
df
```

```python
df['평균'] = df['총합']/5
df
```

```python
df['결과'] = 'Fail'
df
```

```python
df.loc[df['평균']>60,'결과'] = 'Pass'
# 평균이 60 이 넘는 학생들의 결과는 Pass 낮은 학생들은 Fail
df
```

## Column삭제 

```python
df.drop(columns=['총합','평균']) #총합,평균 Column 삭제
```

## Row 삭제

```python
df.drop(index='4번') #4번 학생 데이터 삭제
```

```python
filt = df ['수학']<80 #수학점수가 80점 미만 학생들 필터링
df[filt]
```

## Cell 수정

```python
df.loc['4번','SW특기'] = 'Python' # 4번학생의 SW특기 데이터를 Python 으로 변경 
df
```

```python
df.loc['5번',['학교','SW특기']] = ['능남고등학교','C'] #5번 학생의 학교는 능남고등학교 SW특기는 C로 변경
df
```
